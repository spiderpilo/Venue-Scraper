## Approach 3: Minimal Scraper

````python
import time
import os
import pandas as pd
from src.scraper import scrape_venue_pages
````

````python
INPUT = "data/golden_model_2.csv"
OUTPUT = "data/scraped_cache_train_small_dataset.csv"
````

````python
df = pd.read_csv(INPUT)
````

````python
# df.head(10)
````

````python
df.columns.tolist()
````

````python
def download_to_unique_csv(df, base_filename="data/scraped_cache_train_small_dataset", extension=".csv"):
    # Check if the file already exists
    if not os.path.exists(OUTPUT):
        new_filename = f"{base_filename}{extension}"
    else:
        # If it extists, append a counter & increment until a valid number is found:
        counter = 1
        filename = f"{base_filename}{extension}"
        while os.path.exists(filename):
            filename = f"{base_filename}_{counter}{extension}"
            counter += 1
    df.to_csv(filename, index=False)
    print(f"DataFrame saved as: {filename}")
````

````python
required = [
    "venue_name",
    "Source URL",
    "Business Type",
    "Incentive Category",
    "Experience Category",
    "Incentive Teaser",
    "Full Incentive Description",
]
````

````python
# clean = df.dropna(subset=["Source URL", "Incentive Category", "Full Incentive Description"]).copy()
clean = df.dropna(subset=required).copy()
````

````python
# clean
````

````python
clean = clean[
    clean["Source URL"].astype(str).str.startswith("http")
    & ~clean["Source URL"].astype(str).str.contains(r"\s", regex=True)
]

sample = clean.sample(10, random_state=42)
````

````python
sample["Business Type"]
````

````python
rows = []
for i, row in sample.iterrows():
    name = row["venue_name"]
    url = row["Source URL"]
    btype = row.get("Business Type", "")
    experience_cat = row.get("Experience Category", "")
    incentive_cat = row.get("Incentive Category", "")
    teaser = row.get("Incentive Teaser", "")
    description = row.get("Full Incentive Description")

    print(f"[{i}] scraping {name}: {url}")
    print(f"{btype}, {experience_cat or "Err"}, {incentive_cat or "Err"}, {teaser or "Err"}, {description or "Err"}")
    text = scrape_venue_pages(url, business_type=btype)

    rows.append({
        "venue_name": name,
        "Source URL": url,
        "business_type_gold": btype,
        "experience_category_gold" : experience_cat,
        "incentive_category_gold": incentive_cat,
        "teaser_gold": teaser,
        "description_gold": description,
        "scraped_text": text,
        "text_chars": len(text or ""),
    })
    time.sleep(10)
df_output = pd.DataFrame(rows)
download_to_unique_csv(df_output)
````
